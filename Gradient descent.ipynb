```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('ex1data1.txt',names = ['x','y'])
```

```python
X = data.iloc[:,:1]
y = data.iloc[:,1]
m = len(X)
alpha = 0.01
X.insert(0, 'one', 1, allow_duplicates=True)
n = 1000
```

```python
theta = np.array([[0], [0]])
```

```python
i = 0
cost = []
while i < n:
    
    y_pred = pd.DataFrame(np.dot( X,theta))

    cost.append((1/2*m)*sum((y - y_pred)[0]**2))

    temp0 = theta[0] - ((alpha/m)*(sum((y-y_pred)[0])))
    temp1 = (theta[1] - (alpha/m)*(sum((y - y_pred)[0]*X.iloc[:,1])))
    
    theta[0] = temp0
    theta[1] = temp1
    
    i += 1
```
